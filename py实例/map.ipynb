```python
import matplotlib.pyplot as plt
import random

from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False
```

```python
""" 基础绘图 """

# x,y 轴信息
x = range(40)
y_train = [random.uniform(1,6) for i in x]
y_new = [random.uniform(5,8) for i in x]
# 创建画布
plt.figure(figsize=(10,4),dpi=80)

# 绘制图像
plt.plot(x,y_train,label="上海")
plt.plot(x,y_new,color="r",linestyle="--",label="北京")

# 添加显示x，y轴刻度
x_ticks_label = ["11:%s"%i for i in x]
y_tick = range(30)
plt.xticks(x[::5],x_ticks_label[::5])
plt.yticks(y_tick[::5])

# 添加网格
plt.grid(True,linestyle="--",alpha=0.5)

# 添加图例
plt.legend(loc="best")

# 添加描述信息
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("温度变化图",fontsize=20)

# 显示图像
plt.show()
```

```python
""" 多坐标系绘图 """

# x,y 轴信息
x = range(40)
y_train1 = [random.uniform(1,6) for i in x]
y_train2 = [random.uniform(8,12) for i in x]
# 创建画布
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(10,4),dpi=80)

# 绘制图像
axes[0].plot(x,y_train1,label="上海")
axes[1].plot(x,y_train2,color="r",linestyle="--",label="北京")

# 添加显示x，y轴刻度
x_ticks_label = ["11:%s"%i for i in x]
y_tick = range(30)
axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_tick[::5])
axes[0].set_xticklabels(x_ticks_label[::5])

axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_tick[::5])
axes[1].set_xticklabels(x_ticks_label[::5])


# 添加网格
axes[0].grid(True,linestyle="--",alpha=0.5)
axes[1].grid(True,linestyle="--",alpha=0.5)

# 添加图例
axes[0].legend(loc="best")
axes[1].legend(loc="best")

# 添加描述信息
axes[0].set_xlabel("时间")
axes[0].set_ylabel("温度")
axes[0].set_title("温度变化图",fontsize=20)

axes[1].set_xlabel("时间")
axes[1].set_ylabel("温度")
axes[1].set_title("温度变化图",fontsize=20)

# 显示图像
plt.show()
```

```python
""" 折线图 """
import numpy as np
x = np.linspace(-10,10,1000)
y = np.sin(x)
plt.figure(figsize=(20,8),dpi=60)

plt.plot(x,y)
plt.grid()
plt.show()
```
